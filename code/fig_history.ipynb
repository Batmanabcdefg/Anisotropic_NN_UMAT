```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.prop_cycle'] = plt.cycler('color', plt.cm.inferno(np.linspace(0.2, 0.8, 3)))
fsize=10
pltparams = {'legend.fontsize': 'large',
          'figure.figsize': (fsize*1.75,fsize),
          'font.size'     : 1.4*fsize,
          'axes.labelsize': 1.6*fsize,
          'axes.titlesize': 1.4*fsize,
          'xtick.labelsize': 1.4*fsize,
          'ytick.labelsize': 1.4*fsize,
          'lines.linewidth': 2,
          'lines.markersize': 7,
          'axes.titlepad': 25,
          "mathtext.fontset": 'dejavuserif',
          'axes.labelpad': 5}
plt.rcParams.update(pltparams)
import pickle
```

```python
data = []
a0 = 50
a1 = 10
a2 = 100

nns = ['S112S2_xy_nonconv','S112S2_xy','S112S2_xys_nonconv','S112S2_xys','P1C1_xy','P1C1_xys','P12BC2_xy','P12BC2_sxsy',
       'P12BC2_xyb','P12BC2_bsxsy','P12BC2_xysxsy']
for nn in nns:
    with open('savednet/{name}_history.pkl'.format(name=nn), 'rb') as f:
        history = pickle.load(f)
        loss, sigma, symmetry, convexity, Psi = history
        data.append([loss, sigma, symmetry, convexity, Psi])
```

```python
data1 = data[:4]
```

```python
fig = plt.figure()
gs = fig.add_gridspec(3,5, wspace=0.4, hspace=0.3)

data1 = data[:4]
for i,di in enumerate(data1):
    axi = fig.add_subplot(gs[0,i])
    axi.set(yscale = 'log', xscale='linear')
    axi.plot(di[4], label = '$\mathcal{L}_1$')
    axi.plot(di[1], label = '$\mathcal{L}_2$')
    axi.plot(di[2]+di[3], label = '$\mathcal{L}_3$')
    # axi.plot(di[3], label = '$\mathcal{L}_4$')
    axi.spines['top'].set_visible(False)
    axi.spines['right'].set_visible(False)
    axi.set(xlabel='Iteration')
    if i==0:
        axi.set(ylabel='Fig. 4\nTraining loss')

axi.legend(bbox_to_anchor =(2, 1))

data2 = data[4:6]
for i, di in enumerate(data2):
    axi = fig.add_subplot(gs[1,i])
    axi.set(yscale = 'log', xscale='linear')
    axi.plot(di[4], label = '$\mathcal{L}_1$')
    axi.plot(di[1], label = '$\mathcal{L}_2$')
    axi.plot(di[2]+di[3], label = '$\mathcal{L}_3$')
    # axi.plot(di[3], label = '$\mathcal{L}_4$')
    axi.spines['top'].set_visible(False)
    axi.spines['right'].set_visible(False)
    axi.set(xlabel='Iteration')
    if i==0:
        axi.set(ylabel='Fig. 5\nTraining loss')


data3 = data[6:]
for i, di in enumerate(data3):
    axi = fig.add_subplot(gs[2,i])
    axi.set(yscale = 'log', xscale='linear')
    axi.plot(di[4], label = '$\mathcal{L}_1$')
    axi.plot(di[1], label = '$\mathcal{L}_2$')
    axi.plot(di[2]+di[3], label = '$\mathcal{L}_3$')
    # axi.plot(di[3], label = '$\mathcal{L}_4$')
    axi.spines['top'].set_visible(False)
    axi.spines['right'].set_visible(False)
    axi.set(xlabel='Iteration')
    if i==0:
        axi.set(ylabel='Fig. 6\nTraining loss')

fig.savefig('figs/fig_history.jpg',bbox_inches='tight',dpi=500)
pass
```
