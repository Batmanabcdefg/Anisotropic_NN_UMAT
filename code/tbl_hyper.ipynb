```python
import numpy as np
import pandas as pd
import tensorflow as tf
from misc import preprocessing, normalization, predict
import pickle
```

```python
weights = [[0, 0], 
                [0.050, 0.002], [0.050, 0.004], [0.050, 0.008], 
                [0.075, 0.002], [0.075, 0.004], [0.075, 0.008],
                [0.100, 0.002], [0.100, 0.004], [0.100, 0.008], [0.1, 0.016],
                                                [0.2  , 0.008], [0.2, 0.016],
                                                                        [0.4, 0.032],
                                                                                [0.8, 0.064]]
```

```python
a1vals = np.unique(np.array(weights)[:,0])
a3vals = np.unique(np.array(weights)[:,1])
```

```python
L1trn = pd.DataFrame('x',index=a1vals,columns=a3vals)
L2trn = pd.DataFrame('x',index=a1vals,columns=a3vals)
L3trn = pd.DataFrame('x',index=a1vals,columns=a3vals)


for a1, a3 in weights:
    model_name = 'P1C1_s_a1_{p1}_a3_{p3}'.format(p1=a1,p3=a3)
    dataset_name = 'P1C1_s'
    
    with open('savednet/'+model_name+'_history.pkl', 'rb') as f:
        [total_loss, sigma_loss, symmetry_loss, convexity_loss, Psi_loss] = pickle.load(f)

    ndata = 165
    L1trn.at[a1,a3]=Psi_loss[-1]/ndata*1000
    L2trn.at[a1,a3]=sigma_loss[-1]/ndata*1000
    L3trn.at[a1,a3]=(symmetry_loss[-1] + convexity_loss[-1])/ndata*1000
    pass
```

```python
L1trn.style.set_caption("$\mathcal{{L}}_1$ in training set for $a_1 \in {a1}$ and $a_3 \in {a3}$".format(a1=a1vals, a3=a3vals))
L1trn.style.set_caption("L1 in training set (cols: a1, rows: a3)")
```

```python
L2trn.style.set_caption("$\mathcal{{L}}_2$ in training set for $a_1 \in {a1}$ and $a_3 \in {a3}$".format(a1=a1vals, a3=a3vals))
L2trn.style.set_caption("L2 in training set (cols: a1, rows: a3)")
```

```python
L3trn.style.set_caption("$\mathcal{{L}}_3$ in training set for $a_1 \in {a1}$ and $a_3 \in {a3}$".format(a1=a1vals, a3=a3vals))
L3trn.style.set_caption("L3 in training set (cols: a1, rows: a3)")
```
